## Import data

```python
with open("data/mlops_raw.txt", "r") as file:
    mlops_text = file.read()

mlops_text[:150]
```

## Data models

```python
from pydantic import BaseModel, Field


class Education(BaseModel):
    education_name: str = Field(
        description="Make sure to find this information in the text"
    )
    program_description: str = Field(
        description="A 3-5 sentences summary about the program in a pedagogigcal tone for a student"
    )
    career_options: list[str]
    courses: list[str]
    length: float = Field(description="Total study length of the program")


# language
```

## Create agent

```python
from pydantic_ai import Agent

education_extractor_agent = Agent(
    "openrouter:nvidia/nemotron-nano-12b-v2-vl:free",
    system_prompt="""
    You are an expert in Swedish education focused on Yrkeshögskola (YH). 
    Based on a text about an education, you will extract out relevant information.

    Make sure to use same language as the text.
""",
)

# context - mlops_text
result = await education_extractor_agent.run(mlops_text, output_type=Education)
result.output
```
