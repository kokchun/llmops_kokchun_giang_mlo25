## Clean email json file

Format that MLFlow requires:

- input -> inputs
- {"expected_response": '{'sender_name': 'Erik Lindqvist', 'sender_emai...'}' 

```python
df = df.rename({"input": "inputs"}, axis=1)

df
```

```python
import json
df["expectations"] = df["expectations"].apply(lambda row: {"expected_response": json.dumps(row)})
df
```
